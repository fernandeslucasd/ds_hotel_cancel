```python
d = {'April': 4,
 'August': 8,
 'December': 12,
 'February': 2,
 'January': 1,
 'July': 7,
 'June': 6,
 'March': 3,
 'May': 5,
 'November': 11,
 'October': 10,
 'September': 9}

df['month_arrival_date_index'] = df['month_arrival_date'].map(d)
df['arrival_date'] = pd.to_datetime(dict(year=df.year_arrival_date, month=df.month_arrival_date_index, day=df.day_of_month_arrival_date))
```

```python
cancelyrs = df.groupby('year_arrival_date', as_index=True).agg(cancel_mean=('cancellation','mean'),cancel_std=('cancellation','sem'))
cancelyrs
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])
print(fvalue, pvalue)
```

```python
cancelyrs.plot(y='cancel_mean',yerr='cancel_std')
```

```python
df_cancel = df.groupby('arrival_date', as_index=True).cancellation.agg(cancel_sum='sum')
df_cancel['rolling_avg_cancel'] = df_cancel.cancel_sum.rolling(30).mean()
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(df_cancel.loc[:'2015-12-31'].cancel_sum.values, 
                                df_cancel.loc['2016-01-01':'2017-01-01'].cancel_sum.values, 
                                df_cancel.loc['2017-01-01':].cancel_sum.values)
print(fvalue, pvalue)
```
